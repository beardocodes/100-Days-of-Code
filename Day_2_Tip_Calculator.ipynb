```python
# Greeting message

print("Welcome to the tip calculator.")

#Enter the bill amount

bill = float(input("What was the bill amount?\n$"))

#Enter the tip

tip_percent = int(input("What percentage tip would you like to give? 10, 12, 15?\n"))
tip = (tip_percent / 100) * bill

# Enter number of people

people = int(input("How many people to split the bill?\n"))

# Calculate even bill split

share = (bill + tip) / people
final_share = round(share,2)
message = f"Each person should pay:\n${final_share}"
print(message)
```
